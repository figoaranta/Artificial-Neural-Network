```python
def multilayer_perceptron(input_d,n_layer,n_hidden,activationType):
    print("=========================================================")
    for i in range(0,len(n_hidden)):
        print("neuron in hidden layer",i+1,":",n_hidden[i])
        
    last_layer = None
    last_hidden = None
    layer = [0 for i in range (0,n_layer+1)]
    layer[0] = input_d
    
    for i in range(0,n_layer):
        if activationType[i] == 0:
            layer[i+1] = tf.nn.sigmoid(tf.add(tf.matmul(layer[i], w[i]), b[i]))
        else:
            layer[i+1] = tf.nn.relu(tf.add(tf.matmul(layer[i], w[i]), b[i]))
            
        last_layer = layer[i+1]
        last_hidden = n_hidden[i]

    out_weight = tf.Variable(tf.random_normal([last_hidden, n_output]))
    out_layer = tf.add(tf.matmul(last_layer, out_weight),b[-1])

    return out_layer
```

```python
def trainModel(x_train,y_train,x_test,y_test,n_layer,n_hidden,activationType,learning_constant,number_epochs):
    with tf.Session() as sess:
        neural_network = multilayer_perceptron(X,n_layer,n_hidden,activationType)

    #   Loss Function
        loss_op = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=neural_network,labels=Y))
        optimizer =tf.train.GradientDescentOptimizer(learning_constant).minimize(loss_op)
        init = tf.global_variables_initializer()
        sess.run(init)


        #Training epoch
        for epoch in range(number_epochs):
            sess.run(optimizer, feed_dict={X: x_train, Y:y_train})
            #Display the epoch

        # Test model 
        pred = (neural_network) # Apply softmax to logits
        estimated_class=tf.argmax(pred, 1)#+1e-50-1e-50
        correct_prediction1 = tf.equal(tf.argmax(pred, 1),tf.argmax(y_test, 1))
        accuracy1 = tf.reduce_mean(tf.cast(correct_prediction1, tf.float32))
        
        print("Epoch:",number_epochs,"| Learning Rate:",learning_constant,"| ANN Layer:",n_layer)
        print("Accuracy:",accuracy1.eval({X: x_test}))
        print("=========================================================")
        
        accuracy = accuracy1.eval({X: x_test})
        
        accuracyDict[str(n_layer)+"|"+str(n_hidden)+"|"+str(activationType)+"|"+str(number_epochs)+"|"+str(learning_constant)] = accuracy
        return accuracy
        
```

```python
n_layers = [3]
n_hiddens_list = [[1,2,3,5,10],[1,2,3,5,10],[1,2,3,5,10]]
# activationTypes_list = [[0,1],[0,1],[0,1]]
activationTypes_list = [[0,0,0],[1,1,1]]
n_input = 2
n_output = 2

#Learning parameters
learning_constants = [0.1,0.2,0.5]
number_epochs = [5,10,15,20]
batch_size = [24,32,62]
```

```python
import itertools

X = tf.placeholder("float", [None, n_input])
Y = tf.placeholder("float", [None, n_output])

start = time.time()
for n_layer in n_layers:
    n_hiddens = n_hiddens_list[:int(n_layer)]
    activationTypes = activationTypes_list[:int(n_layer)]
    
    n_hiddens_permutations = (itertools.product(*n_hiddens))
    n_activationTypes_permutations = (itertools.product(*activationTypes))
    
    for learning_constant in learning_constants:
        for epochs in number_epochs:
            for n_activationTypes_permutation in activationTypes_list:
#             for n_activationTypes_permutation in n_activationTypes_permutations:
                for n_hiddens_permutation in n_hiddens_permutations:

                    b = []
                    w = []
                    for i in range (0,len(n_hiddens_permutation)):
                        b.append(tf.Variable(tf.random_normal([n_hiddens_permutation[i]])))
                    b.append(tf.Variable(tf.random_normal([n_output])))

                    w.append(tf.Variable(tf.random_normal([n_input, n_hiddens_permutation[0]])))
                    for i in range (0,len(n_hiddens_permutation)-1):
                        w.append(tf.Variable(tf.random_normal([n_hiddens_permutation[i], n_hiddens_permutation[i+1]])))

            #         print(len(n_hiddens_permutation),list(n_hiddens_permutation),activationType)
                    trainModel(batch_x_train,batch_y_train,batch_x_test,batch_y_test,len(n_hiddens_permutation),list(n_hiddens_permutation),list(n_activationTypes_permutation),learning_constant,epochs)
        
end = time.time()
print("Total run time for parameters tuning:",end - start)
```

```python
n_hidden = [2,10,2]
n_input = 2
n_output = 2
activationType = [0,0,0]

# Best Network parameters
X = tf.placeholder("float", [None, n_input])
Y = tf.placeholder("float", [None, n_output])

#Learning parameters
learning_constant = 0.2
number_epochs = 5
batch_size = 62
fold = 10
b = []
w = []
b.append(tf.Variable(tf.random_normal([n_hidden[0]])))
b.append(tf.Variable(tf.random_normal([n_hidden[1]])))
b.append(tf.Variable(tf.random_normal([n_hidden[2]])))
b.append(tf.Variable(tf.random_normal([n_output])))

w.append(tf.Variable(tf.random_normal([n_input, n_hidden[0]])))
w.append(tf.Variable(tf.random_normal([n_hidden[0], n_hidden[1]])))
w.append(tf.Variable(tf.random_normal([n_hidden[1], n_hidden[2]])))
w3 = tf.Variable(tf.random_normal([n_hidden[2], n_output]))


kf = KFold(n_splits=fold)

accuracy = []
start = time.time()
for train_index, test_index in kf.split(x):
    x_train = (x.iloc[train_index,:])
    y_train = y[train_index]
    x_test = (x.iloc[test_index,:])
    y_test = y[test_index]
    
    accuracy.append(trainModel(x_train,y_train,x_test,y_test,len(n_hidden),n_hidden,activationType,learning_constant,number_epochs))
    
end = time.time()
print("Final Accuracy after 10-k fold cross-validation:",sum(accuracy)/len(accuracy))
print("Total run time for 10-k fold cross-validation:",end - start)
```
